# Otodom.pl Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv("Otodom_Data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df['Title'].duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df['Voivodeship'] = df['Location'].apply(lambda x: x.split(', ')[-1] if pd.notnull(x) else None)

df.head()
```

```python
df['Voivodeship'] = df['Voivodeship'].str.capitalize()
```

```python
df.head()
```

```python
df['City'] = df['Location'].apply(lambda x: x.split(', ')[-2] if pd.notnull(x) else None)

df.head()
```

```python
df['Price'] = pd.to_numeric(df['Price'].str.replace('[ zł,]', '', regex=True), errors='coerce')
```

```python
df['Surface'] = df['Surface'].str.extract('([\d,\.]+)')[0]
```

```python
df.head()
```

```python
df['Surface'] = pd.to_numeric(df['Surface'].str.replace(',', '.'), errors='coerce')
```

```python
df.head()
```

```python
Finishing_Condition = df['Finishing_Condition'].value_counts(dropna=False)

print(Finishing_Condition)
```

```python
Heating = df['Heating'].value_counts(dropna=False)

print(Heating)
```

```python
Parking_Space = df['Parking_Space'].value_counts(dropna=False)

print(Parking_Space)
```

```python
Balcony_Garden_Terrace = df['Balcony_Garden_Terrace'].value_counts(dropna=False)

print(Balcony_Garden_Terrace)
```
